```python
 ## How to establish the connection between SQL Server and Python using SQLAlchemy
conn = sqlalchemy.create_engine(f'mssql+pyodbc:// {socket.gethostname()}/SelormDB?trusted_connection=yes&driver=ODBC Driver 17 for SQL Server')
for fl in os.listdir():
    df = pd.read_csv(fl)
    df.to_sql('sbacovid',con = conn, if_exists = "append", index=False)
```
